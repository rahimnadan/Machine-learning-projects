# Data
The dataSet is downloaded from kaggle and is also present on github on various repositories.

```python
teams= pd.read_csv("teams.csv")
teams
```

```python
teams= teams[["team", "country" , "year", "athletes" , "age" , "prev_medals" , "medals"]]
teams
```

```python
# checking for correlation
teams.corr()["medals"]
```

```python
# visualizing the features with higher correlation to check that either it make a sense or not
import seaborn as sns
sns.lmplot(x="athletes", y= "medals", data= teams, fit_reg= True ,ci=None)
```

```python
# checking for countries that how many medals they have one
teams.plot.hist(y="medals")
```

 In the above figure it is shown that onnly few coutries have won high number of medals and most of the countries have won low number of medals. so this shows that we have some inconsistencies in our data.


```python
# Data cleaning

# looking for missing values
teams[teams.isnull().any(axis=1)]

```

```python
# droping missing values
teams=teams.dropna()
teams
```

# Data Splitting
     Splitting the data into training and testing Sets

```python

train= teams[teams["year"]< 2012].copy()
test= teams[teams["year"] >= 2012].copy()
```

```python
# about 80%
train.shape
```

```python
# about 20%
test.shape
```

```python
# Training the model
from sklearn.linear_model import LinearRegression
reg= LinearRegression()

predictors= ["athletes" , "prev_medals"]
target = "medals"

reg.fit(train[predictors], train[target])


```

```python
prediction = reg.predict(test[predictors])
prediction
```

```python
# Now this is a numpy array which is not our desired result, instead use a column to store the values.
test["prediction"]= prediction
```

```python
test
```

```python
# since the data is in inconsistent form, becuase a team cannot win negative no of medals or half medals. so formatting data.
test.loc[test["prediction"] < 0, "prediction"] = 0  # convert the value less than 0 to 0
test["prediction"]= test["prediction"].round()  # round the numbers into whole number
test
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test["medals"], test["prediction"])
error
```

```python
teams.describe() ["medals"] # if the error value is less than Standard deviation, then this error is acceptable, means our accuracy is good.
```

```python
# Now we have to look the actual medals and predictions for a team
test[test["team"] == "USA"]
```

```python
# We have to find absolute error, i.e (|actual-prediction|)
error= (test["medals"]-prediction).abs()
error
```

```python
# We have to check the average error by team
error_by_team = error.groupby(test["team"]).mean()
error_by_team 
```

```python
# Now we have to find medals by team and check our error ratio between actual medals and predicted medals
medals_by_team = test["medals"].groupby(test["team"]).mean()
medals_by_team
```

```python
error_ratio = error_by_team / medals_by_team
error_ratio
```

it contain some NaN values, it is becuase if a team have won 0 medals and the error is also 0, then it will be a NaN

```python
# we have to only show the non null values
error_ratio[~pd.isnull(error_ratio)]
```

```python
# some of the values are infinite, it is becuase medals won by some teams = 1 and error_by_team= 0, so the ratio results to INF.

# Dealing/removing with infinite values
import numpy as np
error_ratio = error_ratio[np.isfinite(error_ratio)]
error_ratio
```

```python
# Now look at histogram to evaluate our model
error_ratio.plot.hist()
```

```python
# Now sort the values to saw the accuracy 
error_ratio.sort_values()
```
